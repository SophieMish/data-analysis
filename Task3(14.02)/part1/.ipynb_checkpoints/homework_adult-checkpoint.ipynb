```python
import numpy as np
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.stats import zscore
```

```python
#Task 1

class Features(BaseEstimator, TransformerMixin):
    
    def fit(self, x, y=None):
        return self
        
    def transform(self, x, y=None):
        return (x-x.mean(axis=0)) / x.std(axis=0)

    def fit_transform(self, x, y=None):
        return self.fit(x).transform(x)


# class GetNumCols(BaseEstimator, TransformerMixin)
#     def fit(self, x, y=None):
#         return self
        
#     def transform(self, x, y=None):
#         return x[['age', 'hours-per-week']]

#     def fit_transform(self, x, y=None):
#         return self.fit(x).transform(x)
```

```python

class Estimator(BaseEstimator, TransformerMixin):

    def fit(self, x, y=None):
        col_median = x.median(axis=0)
        z = zscore(x,axis=0)
        return col_median, z

    def transform(self, x, y=None):
        columns = x.columns
        col_median, z = self.fit(x)
        x = np.where(z < 3, x, np.nan)
        x = pd.DataFrame(x, columns=columns)
        x.fillna(col_median, inplace=True)
        return x

 
    def fit_transform(self, x, y=None):
        return self.transform(x)
```

**Среди кого больше доля зарабатывающих много(>50K): среди женатых или холостых мужчин (признак marital-status)?**

```python
mask_male =  data['sex']=='Male'
mask_married = data['marital-status'].isin(['Married-civ-spouse','Married-spouse-absent','Married-AF-spouse'])
mask_max_salary = data['salary'] == '>50K'
married_rich = (mask_male & mask_max_salary & mask_married).sum()
single_rich = (mask_male & mask_max_salary & ~mask_married).sum()
married_poor = (mask_male & ~mask_max_salary & mask_married).sum()
single_poor = (mask_male & ~mask_max_salary & ~mask_married).sum()
males = mask_male.sum()
print('Single:',single_rich,'\nRatio of single rich to all males:',round(single_rich/males,2),'\nRatio of single rich vs. single poor:',round(single_rich/single_poor,2))
print('\nMarried:',married_rich,'\nRatio of married rich to all males:',round(married_rich/males,2),'\nRatio of married rich vs. married poor:',round(married_rich/married_poor,2))

```
